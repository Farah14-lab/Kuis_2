```python
#pertama-tama kita import numpy dan pandas, setelah itu kita buka dan tampilkan 100 data pada file dataset_quiz.xlsx 
import numpy as np
import pandas as pd

df = pd.read_excel('data/dataset_quiz.xlsx')

df.head(100)
```

```python
#kita tampilkan informasi detail dataframe berdasarkan baris, kolom, dsb
df.info()
```

```python
#kita kembalikan dan jumlahkan data yang bernilai null berdasarkan nama kolomnya
df.isnull().sum()
```

```python
#kita buat perintah untuk menghapus nilai null pada setiap kolom
modus_ketua = df["id_ketua"].mode()[0]
df["id_ketua"] = df["id_ketua"].replace(np.nan, modus_ketua)

modus_anggota1 = df["id_anggota1"].mode()[0]
df["id_anggota1"] = df["id_anggota1"].replace(np.nan, modus_anggota1)

modus_anggota2 = df["id_anggota2"].mode()[0]
df["id_anggota2"] = df["id_anggota2"].replace(np.nan, modus_anggota2)

modus_anggota3 = df["id_anggota3"].mode()[0]
df["id_anggota3"] = df["id_anggota3"].replace(np.nan, modus_anggota3)

mean_anggota4 = df["id_anggota4"].mean()
df["id_anggota4"] = df["id_anggota4"].replace(np.nan, mean_anggota4)

#menghapus data yang nilainya null pada kondisi seluruh data null dan hanya 1 data bernilai benar
df.drop('id_anggota5', axis=1, inplace=True)
df.head()
```

```python
#menampikkan data dan mengecek data apakah masih terdapat null pada dataframe
df.isnull().sum()
```

```python
#membagi data string menjadi daftar string sesuai jenisnya
from sklearn.model_selection import train_test_split

df_train, df_unseen = train_test_split(df, test_size=0.2, random_state=3, stratify=df['keterlibatan_vltr'])

df_val, df_test = train_test_split(df_unseen, test_size=0.5, random_state=0, stratify=df_unseen['keterlibatan_vltr'])

print(f'========================================================================')
print(f'Total label data asli  : \n{df.keterlibatan_vltr.value_counts()}')
print(f'------------------------------------------------------------------------')
print(f'Total label data valid : \n{df_val.keterlibatan_vltr.value_counts()}')
print(f'------------------------------------------------------------------------')
print(f'Total label data train : \n{df_train.keterlibatan_vltr.value_counts()}')
print(f'------------------------------------------------------------------------')
print(f'Total label data testing : \n{df_test.keterlibatan_vltr.value_counts()}')
print(f'========================================================================')
```

```python
#mengimport label encoder untuk mengkonversi kata menjadi numerik dengan membuat kolom baru
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

df['keterlibatan_vltr']= label_encoder.fit_transform(df['keterlibatan_vltr'])
df['skema']= label_encoder.fit_transform(df['skema'])
df['luaran_hki']= label_encoder.fit_transform(df['luaran_hki'])
df['jenis_luaran_hki']= label_encoder.fit_transform(df['jenis_luaran_hki'])
df['potensi_hki']= label_encoder.fit_transform(df['potensi_hki'])
df['luaran_lain']= label_encoder.fit_transform(df['luaran_lain'])
df['jenis_luaran_lain']= label_encoder.fit_transform(df['jenis_luaran_lain'])
df['status_luaran_lain']= label_encoder.fit_transform(df['status_luaran_lain'])

#melakukan seleksi label berdasarkan tahun
X = df.iloc[:,6:16]
y = df['tahun']
X.head()
```

```python
#menggunakan metode train_test_split dgn membagi dataset menjadi dua bagian yakni bagian yang digunakan untuk training data dan untuk testing data dengan proporsi tertentu
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
#prediksi label set test
y_pred_dt = dt.predict(X_test)
print(y_pred_dt)
```

```python
#mempenghitungkan set accuracy
acc_dt = accuracy_score(y_test,y_pred_dt)
print("Test set accuracy: {:.2f}".format(acc_dt))
print(f"Test set accuracy: {acc_dt}")
```

```python
#menggunakan metode SVC
from sklearn.svm import SVC

svm_poly = SVC(kernel='poly', degree=8)

#model fit
svm_poly.fit(X_train, y_train)
```

```python
#prediksi data test
y_pred_svm_poly = svm_poly.predict(X_test)
print(y_pred_svm_poly)
```

```python
#melakukan evaluasi accuracy testing data
acc_svm_poly= accuracy_score(y_test, y_pred_svm_poly)

print("Test set accuracy : {:.2f}".format(acc_svm_poly))
print(f"Test set accuracy : {acc_svm_poly}")
```

```python
#melakukan tunning hyperparameter berdasarkan hasil klasifikasi pada nomor 3 menggunakan metode Grid Search dan Random Search
from sklearn.model_selection import RepeatedKFold

cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

#melakukan inisisasi model
dt = DecisionTreeClassifier()

#mendefinisikan hyper parameter 
params = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : list(range(5,11))
}

#inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
grid = GridSearchCV(dt, param_grid=params, cv=cv)

#fit berdasarkan grid
%timeit grid.fit(X,y)

#evaluasi berdasarkan score
score = grid.score(X,y)

#cetak hasil evaluasi 
print(f'Hasil evaluasi: {score}')

#cetak konfigurasi hyperparameter
print(f'Konfigurasi hyperparameter : {grid.best_params_}')
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

svm = SVC()

#mendefinisikan hyper parameter 
params = {
    'C' : [0.1, 1, 10],
    'degree' : [3],
    'kernel' : ['linear','poly','rbf']
}

#melakukan inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
grid = GridSearchCV(svm, param_grid=params, cv=cv)

#fit atau latih berdasarkan grid
%timeit grid.fit(X,y)

#evaluasi berdasarkan score
score = grid.score(X,y)

#cetak hasil evaluasi 
print(f'Hasil evaluasi: {score}')

#cetak konfigurasi hyperparameter
print(f'Konfigurasi hyperparameter : {grid.best_params_}')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

#inisisasi model
dt = DecisionTreeClassifier()

#mendefinisikan hyperparameter 
params = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : list(range(5,11))
}

#melakukan inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
grid = RandomizedSearchCV(dt, params, cv=cv)

#fit berdasarkan grid
%timeit grid.fit(X,y)

#evaluasi berdasarkan score
score = grid.score(X,y)

#cetak hasil evaluasi 
print(f'Hasil evaluasi: {score}')

#cetak konfigurasi hyperparameter
print(f'Konfigurasi hyperparameter : {grid.best_params_}')
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV

#model SVM 
svm = SVC()

#mendefinisikan hyperparameter 
params = {
    'C' : [0.1, 1, 10],
    'degree' : [3],
    'kernel' : ['linear','poly','rbf']
}

#inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
grid = RandomizedSearchCV(svm, params, cv=cv)

#fit berdasarkan grid
%timeit grid.fit(X,y)

#evaluasi berdasarkan score
score = grid.score(X,y)

#cetak hasil evaluasi 
print(f'Hasil evaluasi: {score}')

#cetak konfigurasi hyperparameter
print(f'Konfigurasi hyperparameter : {grid.best_params_}')
```
